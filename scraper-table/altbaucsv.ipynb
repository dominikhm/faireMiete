```python
import os
import pandas as pd
import json
import numpy as np
```

```python
data = pd.DataFrame()
for file in os.listdir("./"):
    if file.endswith("data.json"):
        f = open(file, "r") 
        df = json.load(f)
        for key in df.keys():
            try:
                test = pd.DataFrame.from_dict(df[key]).dropna()
                data = pd.concat([data,test],axis=0,ignore_index=True)
            except:
                continue
data = data.drop(labels=0, axis=1).dropna().sort_values('Bezirk')
print(data)
```

```python
data = data[['Bauperiode', 'Bezirk', 'Adresse']]
data
```

```python
data = data[-data.duplicated(subset=['Bezirk', 'Adresse'])]
```
